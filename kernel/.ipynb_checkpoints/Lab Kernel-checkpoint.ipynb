## Generate data

The function MixGauss(means, sigmas, n) generates dataset X,Y where the X is composed of mixed classes, each class being generated according to a Gaussian distribution with given mean and standard deviation. The points in the dataset X are enumerated from 0 to n-1, and Y represents the label of each point.

We start generating two classes of points leaving in a 2D space. The first class is centered in (0,0) with standard deviation 0.5 and the second one in (1,1) with standard deviation 0.3. Each class is made of 15 points.

```python
X, Y = MixGauss(means=[[0,0],[1,1]], sigmas=[0.5, 0.3], n=15)
```

## Study the Kernel matrix

We now build the kernel matrix K of the generated points with Gaussian (RBF) kernel for different values of kernel badwith sigma. 

How does the kernel matrix change with respect to sigma? (Remeber that the kernel function is a measure of similarity between two points)

Can we tell the points comes from two distinct classes? Why?

```python
def plt_hm(sigma):
    K = rbf_kernel(X, X, gamma=1./(2*sigma**2))
    ax = sns.heatmap(K, linewidth=.5, vmin=0, vmax=1)
    plt.show()

interactive_plot = interactive(plt_hm, {'manual': True}, sigma=(.1, 5.0))
output = interactive_plot.children[-1]

interactive_plot
```

## 6 Points Example

We now generate 6 points in a 1D space. The points belong to two classes labled -1 and 1

```python
X, Y = MixGauss(means=[[0],[3]], sigmas=[1, 2.5], n=3)
Y = 2*np.mod(Y, 2)-1
```

Given the set of points we just generated $\{\widehat x_1,\widehat x_2, \dots,\widehat x_6\}$,
the function that the kernel method aim to learn is the following

$$
f(x) = \sum_{i=1}^6 k(\widehat x_i, x) ~ c_i
$$

where $c_i$ are the weights that needs to be learned and the function $k(\widehat x_i, x)$ is the kernel function evaluated at each of the points of the set.

We now study how this function looks like as the badwith of the kernel changes. In particular we are going to consider the kernel function to be the Gaussian kernel

$$
k(a,b) = exp \Bigg\{-\frac{\|a - b\|^2}{2\sigma^2}\Bigg\}
$$


### Before training

First we consider the function before the weights $c_i$ are learned (we give them an initialization value of 0.5 for graphical purposes)


```python
d_min = np.min(X)-2
d_max = np.max(X)+2
disc = np.linspace(d_min,d_max, 5000)

init_val = .5
weights = np.ones((len(X),1)) * init_val

def plot_rbf_fun(sigma):
    plt.figure(figsize=(15, 5))
    plt.ylim(-4,4)        
    K_dx = rbf_kernel(disc.reshape(-1,1), X, gamma=1./(2*sigma**2))
    for i in range(len(X)):
        y = K_dx[:,i] * weights[i]
        plt.plot(disc,y, 'b')
    Yo = K_dx.dot(weights)
    plt.plot(disc, Yo, 'r')
    plt.scatter(X,Y, c='r')
    plt.show()

interactive_plot = interactive(plot_rbf_fun, {'manual': True}, sigma=(.1, 2.0))
output = interactive_plot.children[-1]

interactive_plot
```

As we see the function is the sum of 6 Gaussian functions centerd in the points $\{\widehat x_1,\widehat x_2, \dots,\widehat x_6\}$

What happens as the bandwith varies?

### After training 

Now we consider the function after the weights $c_i$ are learned.

```python
d_min = np.min(X)-.5
d_max = np.max(X)+.5
disc = np.linspace(d_min,d_max, 5000)

def plot_rbf_fun(sigma):
    plt.figure(figsize=(15, 5))
    plt.ylim(-4,4)
    K_dx = rbf_kernel(disc.reshape(-1,1), X, gamma=1./(2*sigma**2))
    KRR = KernelRidge(kernel='rbf', gamma=1./(2*sigma**2), alpha=0)
    KRR.fit(X,Y)
    weights = KRR.dual_coef_
    for i in range(len(X)):
        y = K_dx[:,i] * weights[i]
        plt.plot(disc,y, 'b')
    Yo = K_dx.dot(weights)
    plt.plot(disc, Yo, 'r')
    plt.scatter(X,Y, c='r')
    plt.show()

interactive_plot = interactive(plot_rbf_fun, {'manual': True}, sigma=(.1, 2.0))
output = interactive_plot.children[-1]

interactive_plot
```

## Clustering

We now consider the unsupervised learning problem of clustering.

The dataset we focus on consists of two classes of 2D points.
The points of each class are distributed on a circle. The two circles have the same center but different radius. 

```python
X, Y = make_circles(n_samples=400,noise=.1, factor=.2)
fig, axs = plt.subplots(1,  1)


axs.scatter(X[:, 0], X[:, 1], s=50, c=Y)
axs.set_title('circles dataset')

plt.show()
```

We now compare the standard k-means algortihm for k = 2 with kernel k-means.
What is the difference between the two?
How does sigma affect the solution of kernel k-means?

```python
k=2
def plot_tm_clust(sigma, kernel):
    KerKM = KernelKMeans(n_clusters= k, kernel=kernel, gamma=1./(2*sigma**2))
    out = KerKM.fit_predict(X)
    fig, axs = plt.subplots(1,  2, figsize=(13,5))
    axs[0].scatter(X[:, 0], X[:, 1], s=50, c=Y)
    axs[0].set_title('circles dataset')
    axs[1].scatter(X[:, 0], X[:, 1], s=50, c=out)
    axs[1].set_title('k-means reconstruction')
    plt.show()
    
interactive_plot = interactive(plot_tm_clust, {'manual': True},
                               kernel=['rbf', 'linear'], 
                               sigma=(.01, 1,.05))
output = interactive_plot.children[-1]


interactive_plot
```

### FALKON

We now focus on a real dataset and observe how computationally KRR scale with respect to the number of points

We consider a subset of the SUSY dataset (a high energy phisics binary classification dataset)

```python
n_sub = 5000
# n_sub = 10000
# n_sub = 20000
# n_sub = 50000
# n_sub = 200000


print(f'Selecting a subset of {n_sub} over the SUSY dataset')
X = spio.loadmat('./Susy_small.mat')['X_small'][:n_sub,:]
Y = X[:,0].reshape(-1,1)
Y = (Y-.5)*2
X = X[:,1:]
```

```python
ttp = .8
ntt = X.shape[0]
ntr = int(ntt*ttp)
Xtr = X[:ntr,:]
Ytr = Y[:ntr]
Xts = X[ntr:,:]
Yts = Y[ntr:]
print(f'Training set size ntr = {ntr}')
print(f'Test set size nts = {ntt-ntr}')
print('---')
print(f'Normalizing points according to z-score')
scaler = StandardScaler()
scaler.fit(Xtr)
Xtr = scaler.transform(Xtr)
Xts = scaler.transform(Xts)
```

For the following experiments we fix a sigma and a lambda because we do not focus particularly on the accuracy of the function we learn, but CV procedure can always be applied to improve in that direction

We first test the classical KRR

```python
# For ntr too big IT MAY NOT TERMINATE
lam = 1e-1
sigma = 4
kernel = 'rbf'

tic()
KRR = KernelRidge(kernel=kernel, gamma=1./(2*sigma**2), alpha=lam)
KRR.fit(Xtr,Ytr)
time_KRR = toc()
Ypred = KRR.predict(Xts)

C_error = np.mean(np.sign(Ypred) != np.sign(Yts))
print(f'KRR Classification error = {C_error * 100:.2f}%')
print(f'KRR computational time = {time_KRR:.2f} sec')
```

We now compare with an approximated method dubbed FALKON 

```python
sigma = 4
kernel = lambda A,B: rbf_kernel(A,B, gamma=1./(2*sigma**2))
lam = 1e-6
T = 20
M = 500

tic()
alpha, C = FALKON(Xtr, Ytr, M, kernel, lam, T)
time_falkon = toc()
Ypred = kernel(Xts, C).dot(alpha)
c_err_func = lambda Ypred, Ytrue: np.mean(np.sign(Ypred) != np.sign(Ytrue))
print(f'FALKON Classification error = {c_err_func(Ypred, Yts) * 100:.2f}%')
print(f'FALKON computational time = {time_falkon:.2f} sec')
```

We can see that they achive similar accuracy with the approximated method FALKON faster than standard KRR
